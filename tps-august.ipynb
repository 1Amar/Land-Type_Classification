```python
train_dff = train_df.drop(['id', 'loss'], axis=1)
```

```python
train_dff
```

```python
train_yy = train_df['loss']
```

```python
train_yy
```

```python
N_SPLITS = 10
N_ESTIMATORS = 10000
EARLY_STOPPING_ROUNDS = 200
VERBOSE = 1000
SEED = 299792458
```

```python
features = train_dff.columns.to_list()
#features
```

```python
lgb_params = {
    'objective': 'regression',
    'metric': 'rmse',
    'n_estimators': N_ESTIMATORS,
    'random_state': SEED,
    'learning_rate': 5e-3,
    'subsample': 0.8,
    'subsample_freq': 1,
    'colsample_bytree': 0.6,
    'reg_alpha': 6.4,
    'reg_lambda': 1.8,
    'min_child_weight': 256,
    'min_child_samples': 20,
    'importance_type': 'gain',
    'device': 'gpu',
    'gpu_platform_id': 0,
    'gpu_device_id': 0
}
```

```python
mlp_oof = np.zeros(train_df.shape[0])
lgb_oof = np.zeros(train_df.shape[0])
mlp_pred = np.zeros(test_df.shape[0])
lgb_pred = np.zeros(test_df.shape[0])
```

```python
import pickle
import random
import time

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

from scipy.optimize import minimize

import lightgbm as lgb

from sklearn.linear_model import LinearRegression

from scipy.optimize import minimize

import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_rows = 120
pd.options.display.max_columns = 100

import warnings
warnings.simplefilter('ignore')
```

```python
kf = KFold(n_splits=N_SPLITS, shuffle=True, random_state=SEED)

for fold, (trn_idx, val_idx) in enumerate(kf.split(X=train_dff, y=train_yy)):
    print(f"===== fold {fold} =====")
    X_train, y_train = train_dff.iloc[trn_idx], train_yy.iloc[trn_idx]
    X_valid, y_valid = train_dff.iloc[val_idx], train_yy.iloc[val_idx]
    X_test = test_df[features]
    
    start = time.time()
    model = lgb.LGBMRegressor(**lgb_params)
    model.fit(
        X_train, 
        y_train,
        eval_set=[(X_valid, y_valid)],
        eval_metric='rmse',
        early_stopping_rounds=EARLY_STOPPING_ROUNDS,
        verbose=VERBOSE,
    )
    
    lgb_oof[val_idx] = model.predict(X_valid)
    lgb_pred += model.predict(X_test) / N_SPLITS
    
    elapsed = time.time() - start
    rmse = mean_squared_error(y_valid, lgb_oof[val_idx], squared=False)
    print(f"fold {fold} - lgb rmse: {rmse:.6f}, elapsed time: {elapsed:.2f}sec\n")
```

```python
print(f"oof lgb_rmse = {mean_squared_error(train_yy, lgb_oof, squared=False)}")


np.save("lgb_oof.npy", lgb_oof)
np.save("lgb_pred.npy", lgb_pred)
```
